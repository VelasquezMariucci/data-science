```python
from google.colab import drive
drive.mount('/content/drive')

!pip install --quiet python-dotenv mistralai

import dotenv
import os

# Load your API key from the .env file
!ls "/content/drive/MyDrive/Credentials"
dotenv.load_dotenv('/content/drive/MyDrive/Credentials/mistral.env')
api_key = os.getenv('MISTRAL_API_KEY')
print(api_key)

# Import Mistral client
from mistralai import Mistral

# Initialize the client
model = "mistral-large-latest"
client = Mistral(api_key=api_key)

# Make the chat request
chat_response = client.chat.complete(
    model=model,
    messages=[
        {"role": "user", "content":
         "Design a 3-day trip plan to Osaka, Japan. "
         "Include attractions, hotels, transportation, and expenses."}
    ]
)

# Print the AI response
print(chat_response.choices[0].message["content"])
```
